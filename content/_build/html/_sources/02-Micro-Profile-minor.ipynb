## Chapter 4: Genome Annotation and Functional Overview – Making Sense of the Sequence

The assembled contigs (Or the assembled genome downloaded from NCBI) are still just a long string of nucleotides. Annotation transforms them into a catalogue of genes, each with a predicted function. In this chapter we perform structural and functional annotation with **Prokka**, then generate a comprehensive functional overview using **eggNOG‑mapper**. The result is a standardised GenBank file and a set of tables that feed into all downstream probiotic analyses.

```{Tips}

The default parameters will be used if nothing is specified in the command line. 

```


**Critical outputs:**
- `sample.gbk`: Annotated genome in GenBank format—the master file used by ABRicate, antiSMASH, and visualisation tools.
- `sample.gff`: Gene coordinates in GFF3 format, suitable for genome browsers.
- `sample.faa`: Predicted protein sequences (FASTA), the input for almost all functional tools.
- `sample.ffn`: Nucleotide coding sequences.
- `sample.txt`: Summary statistics: number of CDS, tRNAs, rRNAs, etc.

Prokka assigns gene names using a hierarchical approach: first checking against a curated bacterial protein database, then against Swiss‑Prot, and finally generating hypothetical protein labels for those without confident matches. Review the `.tbl` file to see the evidence for each gene name.

### 4.2 Generating a Genome Circle Plot with Proksee

Visual inspection of the assembled genome helps confirm correct annotation and structural features. **Proksee** (https://proksee.ca) accepts a Prokka‑generated GenBank file and produces a circular map showing GC content, GC skew, CDS, and user‑selected annotation tracks. Alternatively, you can use **CGView** for command‑line generation, but the interactive Proksee interface is more accessible for most users.

```{Note}
- Proksee also provide tools including **Prokka**, and other tools for genomic screening. We will use it in chapter 7 as well.
```

### 4.3 Broad Functional Categorisation with eggNOG‑mapper

While Prokka provides per‑gene annotations, eggNOG‑mapper places each protein into an orthologous group and assigns a COG functional category, KEGG KO, and GO terms. This step is essential for comparing genomes and for downstream pathway reconstruction. (Check more details in Section 5.1)

You can input the 

```bash
emapper.py -i results/annotation/prokka/sample.faa \
           --output results/annotation/eggnog \
           --cpu 12 \
           --itype proteins \
           --tax_scope auto
```

Parse the output to generate a COG category bar plot. For a candidate probiotic, a high proportion of genes in categories G (carbohydrates) and E (amino acids) suggests metabolic versatility, while abundant category V (defense) genes may point to multiple bacteriocin clusters.

### 4.4 Comprehensive Domain Annotation with InterProScan

**InterProScan** integrates Pfam, PROSITE, SMART, and other databases. It provides the deepest view of protein domain architecture and is ideal for identifying adhesion, stress, and regulatory domains.

```bash
interproscan.sh -i results/annotation/prokka/sample.faa \
                -f tsv,gff3,html \
                -o results/annotation/interpro/ \
                -cpu 16 -dp
```
